```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Data types to test
data_types = [np.short, np.intc, np.longlong, np.float32, np.char]
data_type_labels = ["int16", "int32", "float32", "int8"]

# Matrix sizes to test (powers of 2)
matrix_sizes = [2**i for i in range(4, 12)]

# Store timing results
times = {
    "int16"  : [23, 19, 38, 232, 1_748, 13_777, 109_389, 872_294],
    "int32"  : [25, 33, 69, 327, 1_770, 14_047, 112_002, 894_795],
    "float32": [14, 21, 60, 302, 1_774, 14_064, 112_160, 895_813],
    "int8"   : [14, 30, 69, 309, 1_877, 14_835, 112_160, 945_586]
}


# Plot the results
plt.figure(figsize=(10, 6), dpi=450)

for label, data in times.items():
  plt.plot(matrix_sizes, data, label=label)

plt.xlabel("Matrix Size (N x N)")
plt.ylabel("Time (MIcro seconds)")
plt.xscale("log", base=2) # Logarithmic x-axis with base 2
plt.yscale("log", base=2) # Logarithmic y-axis with base 2
plt.title("Matrix Multiplication Time vs. Matrix Size")
plt.xticks(matrix_sizes, [str(s) for s in matrix_sizes]) # Set x-axis ticks at powers of 2
plt.legend()
plt.grid(True)


plt.savefig("matrix_multiplication_time.png")
plt.show()
```
